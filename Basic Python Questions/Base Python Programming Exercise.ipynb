```python
def descriptives(*args):
    sum_n= sum(args)
    count_n= len(args)
    min_n = min(args)
    max_n= max(args)
    return {'sum':sum_n,'count':count_n,'min':min_n,'max':max_n}
```

```python
descriptives(15,20,100,900)
```

```python
def area(name):
    if name== 'square':
        s=int(input('Enter the side of square:'))
        sq_area= s*s
        print('The area of Square is :', sq_area)
    elif name == 'rectangle':
        l= int(input('Enter length of the rectangle:'))
        w=int(input('Enter width of the rectangle:'))
        rect_area= l*w
        print('The area of Rectangle is :', rect_area)
    elif name== 'circle':
        r= int(input('Enter the radius of circle:'))
        pi= 3.14
        circle_area= pi*r**2
        print('The area of Circle is:',circle_area)
```

```python
area('circle')
```

```python
area('rectangle')
```

```python
area('square')
```

```python
l1 = ['January', 'February', 'March', 'May', 'June', 'September', 'December']
l2 = ['January', 'February', 'April', 'June', 'October', 'December']
def func(l1,l2):
    l1=set(l1)
    l2=set(l2)
    output = {}
    output['Matched']= list(l1.intersection(l2))
    output['Only in l1'] = list(l1-l2)
    output['Only in l2'] = list(l2-l1)
    return output
```

```python
func(l1,l2)
```

```python
def prime_check():
    n= int(input('Enter number to check:'))
    a=0
    for i in range(2,n):
        if n%i == 0:
            a +=1
        else:
            pass
    if a ==0 :
        print(n,'is prime')
    else:
        print(n,'is prime')
    
```

```python
prime_check()
```

```python
sentence = input('Enter any sentence : ')
sentence = list(sentence)                   # it will dissociate the sentence with all letters
upper = 0
lower = 0
for i in sentence:
    if i.isupper():
        upper = upper+1   
    elif i.islower():
        lower = lower+1
    else:
        pass
print('UPPER CASE:',upper,'LOWER CASE:',lower)
```
