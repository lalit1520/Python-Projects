### Tic Tac Toe

```python
board = [' ' for i in range(9)]

def print_board():
    for i in range(0,9,3):
        print(board[i] + ' | ' + board[i+1] + ' | ' + board[i+2])
        if i < 6:
            print("---------")

def winner(player):
    wins = [
        (0,1,2), (3,4,5), (6,7,8),  # rows
        (0,3,6), (1,4,7), (2,5,8),  # columns
        (0,4,8), (2,4,6)            # diagonals
    ]
    for a,b,c in wins:
        if board[a] == board[b] == board[c] == player:
            return True
    return False

current = 'X'

while True:
    print_board()
    move = int(input(f"Player {current}, enter a position (1â€“9): ")) - 1
    
    if board[move] != ' ':
        print('Spot taken! Try again.')
        continue

    board[move] = current

    if winner(current):
        print_board()
        print(f"ðŸŽ‰ Player {current} wins!")
        break

    if ' ' not in board:
        print_board()
        print("It's a Tie")
        break
        
    current = '0' if current == 'X' else 'X'
```
