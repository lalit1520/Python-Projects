### Guessing Game

```python
number = random.randint(1,100)
attempts = 1
guess = int(input('Guess a Number between 1 and 100: '))

while guess != number:
    if guess > number:
        print(f'Try Guessing Lower')
    else:
        print(f'Try Guessing Higher')
    guess = int(input('Guess a Number between 1 and 100: '))
    attempts += 1
else:
    print()
    print(f'Correct!! The Number was {number}')
    print(f'You guessed it correctly in {attempts} tries')
```

### Rockâ€“Paperâ€“Scissors

```python
import random 

choices = ["rock", "paper", "scissors"]
game_count = 0
user_won = 0 
i_won = 0
ties = 0

while True:
    user = input("Rock, paper, or scissors? (or 'quit'): ").lower()
    if user == 'quit':
        print(f"""
        Total Games: {game_count}
        You won {user_won} games
        I won {i_won} games
        Ties - {ties}
        """)
        break
        
    if user not in choices:
        print('Please Choose Valid Choice! ðŸ˜Š')
        continue

    python_choice = random.choice(choices)
    print("I chose:", python_choice)
    
    if user == python_choice:
        print("                   It's a Tie! ðŸ˜›")
        ties += 1
    elif (user=='rock' and python_choice =='scissors') or \
         (user == 'paper' and python_choice == 'rock') or \
         (user == 'scissors' and python_choice == 'paper'):
        print('                   You Won!! ðŸ˜Œ')
        user_won += 1
        game_count += 1
    else:
        print('                   You Lost! ðŸ˜Ž')
        i_won += 1
        game_count += 1
```
