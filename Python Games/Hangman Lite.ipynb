### Word Guess (Hangman Lite)

```python
words = ["ananya", "katrina", "aneet", "shraddha", "kriti",'alia','kiara','emma','dakota','sadie', 'shailene','kate','zendaya']
words = [i.upper() for i in words]
word = random.choice(words)

display = ['_'] * len(word)
attempts = 5

letters = [i for i in word]
letters = sorted(set(letters))
# print(display)
# print(letters)
# print(word)

while attempts > 0:
    print('Word: ', ' '.join(display))
    guess = input('Guess the actress:').upper()
    
    if guess in letters:
        for index,char in enumerate(word):
            if char == guess:
                display[index] = guess
    else:
        attempts -= 1
        print(f'Wrong guess! Attempts left : {attempts}')

    if '_' not in display:
        print(f'You guessed it: {word}')
        break

if attempts == 0:
    print(f'You lost! The word was: {word}')
```
